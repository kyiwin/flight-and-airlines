# Airline Safety dataset

```python
import pandas as pd
airline_safety = pd.read_csv('airline-safety.csv', index_col = 0)
airline_safety
```

```python
airline_safety.columns
```

```python
airline_safety_ry = airline_safety.drop(['incidents_85_99', 'fatal_accidents_85_99', 'fatalities_85_99'], axis = 1)
airline_safety_ry
```

```python
writer = pd.ExcelWriter('airline_safety-1.xlsx')
airline_safety.to_excel(writer,'85-14')
airline_safety_ry.to_excel(writer,'00-14')
writer.save()
```

```python
airline_safety_ry_A= airline_safety_ry.loc[(airline_safety_ry['incidents_00_14']== 0) & (airline_safety_ry['fatal_accidents_00_14']== 0) & (airline_safety_ry['fatalities_00_14']== 0)]
airline_safety_ry_A                                                                                                          
```

```python
airline_safety_ry_B= airline_safety_ry.loc[(airline_safety_ry['fatal_accidents_00_14']== 0) & (airline_safety_ry['fatalities_00_14']== 0)]
airline_safety_ry_B       
```

```python
airline_safety_ry_C= airline_safety_ry.loc[(airline_safety_ry['fatalities_00_14'] < 10) & (airline_safety_ry['fatalities_00_14'] != 0)]
airline_safety_ry_C
```

```python
airline_safety_ry_D= airline_safety_ry.loc[(airline_safety_ry['fatalities_00_14'] < 50) & (airline_safety_ry['fatalities_00_14'] != 0)]
airline_safety_ry_D
```

```python
airline_safety_ry_E= airline_safety_ry.loc[(airline_safety_ry['fatalities_00_14'] < 100) & (airline_safety_ry['fatalities_00_14'] != 0)]
airline_safety_ry_E
```

```python
airline_safety_ry_F= airline_safety_ry.loc[(airline_safety_ry['fatalities_00_14'] > 100) & (airline_safety_ry['fatalities_00_14'] != 0)]
airline_safety_ry_F
```

```python
writer = pd.ExcelWriter('airline_safety-2.xlsx')
airline_safety_ry_A.to_excel(writer,'A-safety')
airline_safety_ry_B.to_excel(writer,'B-safety')
airline_safety_ry_C.to_excel(writer,'C-safety')
airline_safety_ry_D.to_excel(writer,'D-safety')
airline_safety_ry_E.to_excel(writer,'E-safety')
airline_safety_ry_F.to_excel(writer,'F-safety')
writer.save()
```
